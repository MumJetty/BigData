#### 3.2.1.3 세트

세트는 수학의 집합을 구현한 자료형.  
세트에 저장된 값들은 순서가 없기 때문에 인덱싱과 슬라이싱 불가.  
대신 중복 값 저장이 불가능하다는 특징을 활용해 중복 제거용으로 사용하거나  
교집합, 합집합, 차집합과 같은 집합 개념이 필요한 경우 사용

>사용법
* {값1, 값2}
* set(반복가능객체)
<br>

다만 빈 세트 사용 시에는 중괄호를 사용할 수 없음. (이는 딕셔너리)
빈 세트 생성 시 set 함수 이용

```python
# Jaccard Similarity

string1 = '나는 파이썬을 좋아합니다'
string2 = '나는 파이썬을 정말 좋아합니다'

splitted_string1 = string1.split()
splitted_string2 = string2.split()

print(splitted_string1)
print(splitted_string2)

splitted_string1_set = set(splitted_string1)
splitted_string2_set = set(splitted_string2)

print(splitted_string1_set.intersection(splitted_string2_set))
print(splitted_string1_set.union(splitted_string2_set))

# 인터섹션: 교집합 (세트의 메서드, 순서없음)
# 유니온: 합집합 (세트의 메서드, 순서없음)

print(
    len(splitted_string1_set.intersection(splitted_string2_set))
    / len(splitted_string1_set.union(splitted_string2_set))
)

# 문장의 유사도 계산한 것 

```

#### 3.2.1.4 딕셔너리

딕셔너리는 말 그대로 사전을 의미.  
파이썬에서는 '키'와 '값'을 '단어'와 '단어의 의미'처럼 사용.  



> 사용법
1. dict = {키1: 값1, 키2: 값2, ...}
2. dict = dict(키1=값1, 키2=값2)
3. dict = dict([(키1, 값1), (키2, 값2)]) 
4. dict = dict({키1: 값1, 키2: 값2}) 
5. dict = dict(zip([키1, 키2], [값1, 값2]))

```python
print({'이름': '홍길동', '나이': 16, '직업': '학생'})

print(dict(이름='홍길동', 나이=16, 직업='학생'))
# 키 값이 문자열 처리가 되어있지 않다. 딕셔너리 안에다 변수 만든 것. 변수명이 키값으로 들어간 것. 그래서 따옴표 안씀.

print(dict([('이름', '홍길동'), ('나이', 16), ('직업', '학생')]))

print(dict({'이름': '홍길동', '나이': 16, '직업': '학생'}))

print(dict(zip(['이름', '나이', '직업'], ['홍길동', 16, '학생'])))
# 서로 다른 반복 가능한 객체 묶어줌
# zip(a,b) 외에 zip(a,b,c)도 되고 몇개가 와도 상관없다.

# 1번 3번 5번 많이 씀
```

if-else

if-else문은 조건식을 만족하는 경우와 만족하지 않는 경우를 
구분할 필요가 있을 시 사용.

사용법
* if 조건식:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식의 결과가 False일 때 실행문

조건식의 결과가 True인 경우와 False인 경우로 구분하여 코드 작성 가능.

if-elif

if-elif문은 특정 조건에 따라 3가지 이상으로 구분해야 할 때 사용.
여기서 elif는 else if

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문

if-elif-else

if-elif문은 특정 조건에 따라 구분 후 위의 조건을 만족하는 않는  
모든 경우에 대해 실행을 할 때 사용.

사용법
* if 조건식1:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 True일 때 실행문  
  elif 조건식2:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1의 결과가 False이고 조건식2의 결과가 True일 때 실행문.  
  elif 조건식3:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2의 결과가 False이고 조건식 3의 결과가 True일 때 실행문  
  else:
<br>&nbsp;&nbsp;&nbsp;&nbsp;조건식1, 2, 3이 모두 False일 때 실행될 문장.
